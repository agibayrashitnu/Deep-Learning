```python
#Reading the dataset using Pandas
dataset = pd.read_csv("Churn_Modelling.csv")
```

```python
dataset.head()
```

```python
# Selecting Dependent and Intependent Variables
x = dataset.drop(labels = ['RowNumber', 'CustomerId', 'Surname', 'Exited'], axis = 1)
y = dataset.Exited
```

```python
x.head(10)
```

```python
y.head()
```

```python
# Encoding the categorical data
from sklearn.preprocessing import LabelEncoder
```

```python
label_1 = LabelEncoder()
x.Geography = label_1.fit_transform(x.Geography)
```

```python
x.head()
```

```python
label_2 = LabelEncoder()
x.Gender = label_2.fit_transform(x.Gender)
```

```python
x.head()
```

```python
# Avoiding the dummy variables trap
x = pd.get_dummies(x, drop_first = True, columns= ["Geography"])
x.head()
```

```python
# Splitting the data into train and test split
from sklearn.model_selection import train_test_split

x_train, x_test , y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling Process

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Building the Model

```python
# Creating an object ( Innitializin ANN)

model = tf.keras.models.Sequential()
```

```python
#Adding input layer and first hidden layer

#1) Number of units have to be specified units = 6 ( usually average of input and output (11+1)/2 = 6(just trick))
#2)Activation Function = ReLU
#3) input dimension = 11 ( number of columns that is being taken as input i.e independent variables)

model.add(tf.keras.layers.Dense(units = 6, activation = 'relu', input_dim = 11))
```

```python
x.shape
```

```python
# Adding Second hidden layer
model.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
```

```python
#Output Layer
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
#Compiling the model
model.compile(optimizer = 'adam', loss= 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#Params in first layer is equal to units* (input dim+1)
# Params in second layer = units* (units+1)
# params in last layer = units*(units+1)
# Total params = sum of all params
model.summary()
```

# Training the Model

```python
# x_train is numpy array and y_train is a vector, so we convert vector to numpy array
# batch size = 10 , epochs = 20 means we are passing the data into model 20 times and we pass it in batch of 10
model.fit(x_train,y_train.to_numpy(), batch_size = 10, epochs = 20)
```

# Model Evaluation and Prediction

```python
# Evaluation of the model
test_loss , test_acc = model.evaluate(x_test,y_test.to_numpy())
```

```python
print(f"Test accuracy is: ", test_acc)
```

```python
y_pred = np.argmax(model.predict(x_test), axis=1)
```
